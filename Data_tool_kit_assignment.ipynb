2) How does broadcasting work in NumPy?
- when we are doing the airthematic operations on 2 arrays if the dimensions or shapes or sizes of the arrays does not match then by the help of broadcasting which is implicitly done in numpy library by python to perform the actions on arrays of different shapes.
- to perform the operations on two different dimensional arrays it will check the 2 rules
- **rule 1** : rule-1 it will make the same dimension of small array as large array. and then from left side it will padd the 1's to the small array.
- **rule 2**: it will compare the left sizes of the both arrays then increases the value to 1 to the small array from the large array.

-  eg:  

```python
# example of braodcasting

import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([[23],[24],[25]])
print("dimension of arr1 is ",arr1.ndim)
print("dimension of arr2 is ",arr2.ndim)
print(arr1.shape)
print(arr2.shape)
'''arr1 is 1 dimesnional and arr2 is 2 dimensional here broadcasting comes into the picture
arr1 will become--> (1,3) --- here python by broadcasting virtually changes this value to (3,3)
arr2 is (3,1)
'''

arr1+arr2
```

4)  Explain the use of the groupby() method in Pandas?
- Group by() is used to segregate the dataframe into groups to perform the operations on it and combines the result effectively.
- it is based on based on split-apply-comibe method first it will split the dataframe in to the groups then applies the any user defined function to the groups then the returned outputs from each group are merged into a new DataFrame or Series.

```python
# example of group by usage in data frame.
import pandas as pd

data={"Names":["MOULALI","IMRAN","FAZAL","ASIF","ALI","MOULALI","IMRAN","ALI"],
"Marks":[98,97,87,67,89,89,45,90]}

df=pd.DataFrame(data)
df["total_marks"]=df.groupby("Names")["Marks"].transform("sum")
df.drop_duplicates(subset=['Names'],keep="first")
# step1---> splitting the data as names.
# step 2---> applies sum() function on the marks column of each group

```

7) What is a heatmap, and when should it be used?

- A heatmap is a color-coded grid which helps us to unterstand the data distributions and it is a two-dimensional technique it represents the data as colors in the form of grid or matrix form. usually for higher values warm like colors will be allocates(yellow or red). for low level values cooler color that is blue will be allocated.
```python
# example of heat map
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset("flights")


flight_data=df.pivot(index="year",columns="month",values="passengers")

sns.heatmap(flight_data,annot=True,fmt="d",cmap="coolwarm")
plt.xlabel("month")
plt.ylabel("year")
plt.title("passengers heat map")
plt.show()
```

21) Why is NumPys array slicing faster than Pythons list slicing?
- NumPy's array slicing is faster than Python's list slicing due to differences in how they store and access data.

- Memory Layout
 - NumPy Arrays: Store elements in contiguous memory blocks, allowing for efficient access and manipulation.

 - Python Lists: Store references to objects scattered in memory, leading to less efficient access.
- Implementation Efficiency
 - NumPy: Operations are implemented in C, enabling low-level optimizations and faster execution.

 - Python Lists: Operations are managed by Python's interpreter, introducing more overhead.
